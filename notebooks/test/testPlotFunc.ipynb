# Test Plot Function
Este notebook es un test de integracion de la funcion para graficar escrita en el modulo 'plotUtils.py'

```python
import sys
sys.path.append('../../scripts')
from plotUtils import plotBtcPrice

import requests 
import pandas as pd
import matplotlib.pyplot as plt

# Descarga de datos (ultimos 30 dias)
url = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart'
params = {'vs_currency': 'usd', 'days': '30'}
response = requests.get(url, params=params).json()

# Convertir en DataFrame
prices = pd.DataFrame(response['prices'], columns=['timestamp', 'price'])
prices['timestamp'] = pd.to_datetime(prices['timestamp'], unit='ms')

# Uso la funcion
plotBtcPrice(prices, title='Prueba funcion BTC Price')
```
